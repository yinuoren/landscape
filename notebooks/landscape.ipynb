```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
res = np.load("../results_diag/vgg_gd_rand_4800_r50/landscape.npz")
x = res["x"]
train_loss = res["train_loss"]
train_acc = res["train_acc"]
test_loss = res["test_loss"]
test_acc = res["test_acc"]
```

```python
plt.figure(figsize=(6.5,5))
plt.plot(x, train_loss, linewidth=3.5, color="tab:blue", label="loss")
plt.xticks([-1.0,-0.5,0,0.5,1.0],fontsize=20)
plt.yticks([0,1.5,3.0], fontsize=20)
plt.legend(fontsize=20, loc="upper left")
#plt.savefig("densenet_gd_loss", bbox_inches="tight")
plt.show()
```

```python
# second finite differential
n = len(train_loss)
train_loss_d2 = train_loss[0:n-2] + train_loss[2:n] - 2*train_loss[1:n-1]
plt.figure(figsize=(6.5,5))
plt.plot(x[1:n-1], train_loss_d2, linewidth=3.5, color="tab:orange", label="2nd finite difference")
plt.xlabel("t", fontsize=24)
plt.xticks([-1,-0.5,0,0.5,1], fontsize=20)
plt.yticks([2.14e-3, 2.18e-3, 2.23e-3], fontsize=20)
plt.legend(fontsize=20, loc="lower left")
plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
#plt.savefig("densenet_gd_hess", bbox_inches="tight")
plt.show()
```
